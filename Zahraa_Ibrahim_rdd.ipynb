###### Q2: list all the types of studies in the dataset along with the frequencies of each type

####### Split the pipes

```
myRDD1 = myRDD1.map(lambda x: (x.split("|")))
myRDD1.take(2)
```

#### working with RDD implementation

```
from operator import add
# extract the 5th column at index.
Type_RDD = myRDD1.map(lambda x: (x[5],1))
# Use the reduceByKey (add) function to add the keys together.
Type_Frequency = Type_RDD.reduceByKey(add)
# We order the results in descending order by using the sortBy function.
Type_Frequency = Type_Frequency.sortBy(lambda x: x[1], ascending =False)

# print the results
print (" list all the types of studies in the dataset along with the frequencies of each type :")
print ("         Type\t\t    count")
Type_Frequency.take(5)
```

#### Q3: The top 5 conditions with their frequencies.

```
from operator import add
# extract the 7th column at index
Condition_RDD = myRDD1.map(lambda x: (x[7]))
# run transformation operation to get multiple elements
Condition_RDD = Condition_RDD.flatMap(lambda x: x.split(","))
# filter the rows with null values 
Condition_filter = Condition_RDD.filter(lambda row: row!= "")
# a pair_RDD create from list conditions into a key_value
Condition_filter = Condition_filter.map(lambda s:(s,1))
# Use the reduceByKey (add) function to add the keys together 
Condition_filter = Condition_filter.reduceByKey(add)
# We order the results in descending order by using the sortBy function
Condition_Frequency = Condition_filter.sortBy(lambda x: x[1], ascending =False)
# print the results
print (" The top 5 conditions with their frequencies :")
print ("Conditions\t\tcount")
Condition_Frequency.take(5)
```

#### Q4: The 10 most common sponsors that are not pharmaceutical companies.

```
pharma = sc.textFile("/FileStore/tables/" + pharma_csv_file)
```

####run the RDD implemantation

```
# Remove the header from the pharma dataset.
header = pharma.first()
pharma = pharma.filter(lambda row: row!= header)
# Use the map transformation to split the pipe with a comma.
pharmaRDD = pharma.map(lambda x: x.split('","'))
# a pair_RDD create from list parent_comany into a key_value.
pharma_RDD=pharmaRDD.map(lambda x: x[1])
Pharma_RDD = pharma_RDD.map(lambda x: (x,1))
# Use the reduceByKey (add) function to add the keys together.
Pharma_RDD = Pharma_RDD.reduceByKey(lambda a,b: a+b)
# Remove the header from the clinicaltrial_2021 dataset.
Header = myRDD1.first()
myRDD1 = myRDD1.filter(lambda row: row!= Header)
# a pair_RDD create from list sponsor into a key_value and the reduceByKey (add) function to add the keys together.
sponsor_RDD = myRDD1.map(lambda s: s[1]).map(lambda s: (s,1)).reduceByKey(add)
# We get the difference in pharmaceutical companies that mismatch between sponsor and parent company.
common_sponsor_not_pharametical = sponsor_RDD.subtractByKey(Pharma_RDD).sortBy(lambda x: (x[1]),ascending =False)
# print the results
print (" The 10 most common sponsors that are not pharmaceutical companies :")
print ("             sponsors\t\t            count")
common_sponsor_not_pharametical.take(10)
```

###### Q5: The number of completed studies each month in 2021

```
from operator import add
# extract index 2nd and 4th, then filter status = completed.
studies_RDD = myRDD1.map(lambda x: [x[i] for i in [2,4]]).filter(lambda x: x[0]== "Completed")
# Split the completion column into multiple columns and use the index value of -1. To acces the second to last item.
studies_RDD = studies_RDD.map(lambda x: x[1].split(" ")).filter(lambda x: x[-1] == Year)
# Create a pair of RDD by using a key value.
studies_RDD =studies_RDD.map(lambda x: x[0]).map(lambda r:(r,1))
# count each month's frequencies
studies_RDD = studies_RDD.reduceByKey(lambda a,b: a+b)
# print the results
print (" The number of completed studies each month in 2021 :")
print ("Months\t\tcount")
studies_RDD.take(12)
```

###### sort by months names

```
def sortFunction(arange):
    if arange[0] == "Jan":
        result = (1,arange)
    elif arange[0] == "Feb":
        result = (2,arange)
    elif arange[0] == "Mar":
        result = (3,arange)
    elif arange[0] == "Apr":
        result = (4,arange)
    elif arange[0] == "May":
        result = (5,arange)
    elif arange[0] == "Jun":
        result = (6,arange)
    elif arange[0] == "Jul":
        result = (7,arange)
    elif arange[0] == "Aug":
        result = (8,arange)
    elif arange[0] == "Sep":
        result = (9,arange)
    elif arange[0] == "Oct":
        result = (10,arange)
    elif arange[0] == "Nov":
        result = (11,arange)
    else:
        result = (12,arange)
    return result
completed_study = studies_RDD.map(lambda arange: sortFunction(arange)).sortBy(lambda arange: arange[0], True).map(lambda arange: arange[1])
completed_study.collect()
```

##### Data visualisation by using the bokeh library

```
Months = completed_study.map(lambda x: x[0]).collect()[:10]
count = completed_study.map(lambda x: x[1]).collect()[:10]
```

```
from bokeh.plotting import figure, output_file, show
from bokeh.embed import file_html
from bokeh.resources import CDN
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral10

my_data = {
    "Months": Months,
    "count": count
}
fig = figure(x_range= Months, 
           height=300, 
           title="Number of of completed studies each month in 2021",
           toolbar_location=None, 
           tools="",x_axis_label="MONTHS",y_axis_label= "NUMBER_OF_STUDIES")
fig.add_tools(HoverTool(tooltips = [("Number of studies ", "$y")]))

fig.vbar(x = Months, top = count, width = 0.8, color = Spectral10)
fig.title.align = "center"
fig.xgrid.grid_line_color = None
fig.y_range.start = 0
fig.y_range.end = 1400
fig.title.text_color = "black"
fig.axis.axis_label_text_font_style = "bold"
html = file_html(fig, CDN, "my_data")
displayHTML(html)
```

#### Find the top 5 most common agency action frequencies for pharmaceutical companies.

```
Agency_RDD = pharma.map(lambda x: x.split('","')).map(lambda x: [x[i] for i in [1,12]]).filter(lambda x: x[1]== "agency action")
parent_rdd = Agency_RDD.map(lambda x : x[0]).map(lambda x:(x,1)).reduceByKey(lambda a,b: a+b)
pharma_action_agency = parent_rdd.sortBy(lambda x: (x[1]),ascending =False)
pharma_action_agency.take(5)
```
